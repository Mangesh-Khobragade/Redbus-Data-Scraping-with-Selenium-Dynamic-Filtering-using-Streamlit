#KERALA

```python
driver=webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
time.sleep(3)
driver.maximize_window()
```

```python
#retrieve bus route link
wait=WebDriverWait(driver,20)

def kerala_link_route(path):
    LINK_KERALA=[]
    ROUTE_KERALA=[]

    #retrieve the route link
    for i in range(1,3):
        paths=driver.find_elements(By.XPATH,path)

        for links in paths:
            d=links.get_attribute("href")
            LINK_KERALA.append(d)
        for routes in paths:
            ROUTE_KERALA.append(routes.text)
        try:
            pagination=wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            next_button=pagination.find_element(By.XPATH,f'//div[@class="DC_117_pageTabs" and text()={i+1}]')
            time.sleep(3)
            next_button.click()
        except NoSuchElementException:
            print(f"No more pages at step{i}")
            break
    return LINK_KERALA,ROUTE_KERALA
LINK_KERALA,ROUTE_KERALA=kerala_link_route("//a[@class='route']")
```

```python
df_k =pd.DataFrame({"route_name":ROUTE_KERALA,"route_link":LINK_KERALA})
df_k
```

```python
path=r"C:/Users/HP/Desktop/csv files/df_k.csv"
df_k.to_csv(path,index=False)
```
