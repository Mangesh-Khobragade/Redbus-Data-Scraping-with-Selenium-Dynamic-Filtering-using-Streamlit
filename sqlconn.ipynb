```python
import pandas as pd
import numpy as np
import mysql.connector

```

```python
#csv to dataframe
df_buses_1=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_1.csv")
df_buses_2=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_2.csv")
df_buses_3=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_3.csv")
df_buses_4=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_4.csv")
df_buses_5=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_5.csv")
df_buses_6=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_6.csv")
df_buses_7=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_7.csv")
df_buses_8=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_8.csv")
df_buses_9=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_9.csv")
df_buses_10=pd.read_csv("C:/Users/HP/Desktop/csv files/datails.csv/df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6, df_buses_7,df_buses_8,df_buses_9,df_buses_10], ignore_index=True)
Final_df
```

```python
#DATA CLEANING
Final_df.info()
```

```python
#CONVERT PRICES TO NUMERIC
Final_df["Prices"]=Final_df["Prices"].str.replace("INR","")
Final_df["Prices"].fillna(0)
Final_df["Prices"]=Final_df["Prices"].astype("float")
```

```python
#CONVERT RATINGS TO NUMERIC
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"].fillna(0)
Final_df["Ratings"]=pd.to_numeric(Final_df ["Ratings"],errors="coerce")
```

```python
#info after the datatype change
Final_df.info()
```

```python
#replacing the nan values
Final_df=Final_df.drop(Final_df[Final_df['Prices']==0].index)
Final_df=Final_df.drop(Final_df[Final_df['Ratings']==0].index)
Final_df=Final_df.dropna(subset=['Seats_Available'])
Final_df=Final_df.dropna(subset=['Bus_types'])
```

```python
Final_df['Start_Time'] = pd.to_datetime(Final_df['Start_Time'], format='%H:%M').dt.time
Final_df['End_Time'] = pd.to_datetime(Final_df['End_Time'], format='%H:%M').dt.time
```

```python
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.replace("Seats_Available","")
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.strip()
Final_df["Seats_Available"]=Final_df["Seats_Available"].str.split().str[0]
Final_df["Seats_Available"]=pd.to_numeric(Final_df ["Seats_Available"],errors="coerce")
```
